# WORKING WITH TUPLES

- Tuples are collections like lists, but immutable
- Tuples use the same indexing as list; positive indexing from 0 to access items from the beginning and negative for accessing items from the end

##### **Syntax**
```
(item1, item2, item3, item4, ...)
```

````python
week_days = ('Sun', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat')
````

##### Tuple is immutable
- Once created, we cannot modify it; cannot delete an item, change an item, add an item, sort items, reverse items

### `len(tuple)` function
- Counts the number of items in the tuple and returns the count


````python
len(week_days)
````

### Unpacking items from a tuple
- We can assign each item from a tuple to a separate variable using assignment operator

````python
week_days
````

````python
d1, d2, d3, d4, d5, d6, d7 = week_days
````

````python
d1
````

````python
d2
````

````python
d3
````

````python
d7
````

### Indexing

````python
week_days[0]
````

````python
week_days[-1]
````

**Note:** Unpacking applies to lists as well

````python
we1, we2 = ['Sat', 'Sun']
print(we1)
````

### Indexing applies to tuple and list

````python
week_days
````

##### Syntax
```
tuple[start:end:step]
list[start:end:step]
```

````python
week_days[1:-1]
````

````python
week_days[::2]
````

````python
week_days[1::2]
````

````python
week_days[::3]
````

````python
week_days[::-1]    # reverse order
````

### Two functions to convert between list and tuple: list() and tuple()

````python
type(week_days)
````

````python
week_days_list = list(week_days)
type(week_days_list)
````

````python
week_days_list
````

````python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
months_tuple = tuple(months)
months_tuple
````
