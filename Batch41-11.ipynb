# WORKING WITH FILES

### Two types of files
1. Text files
2. Binary files

### Text files
- A text file contains a sequence of lines
- And each line contains characters and ends with a new line (\n) character
- Examples: txt, html, xml, json, csv, log, etc.

### Binary files
- A binary stores a sequence of bytes; may contain parts of text
- Not a text file
- Example: image, video, audio, compressed, database files, etc.

### Sequence of file operations
1. Open a file - open() function
2. Read from the file or write to the file - read() / write() methods
3. Close the file - close() method

**Note:** Always close the file; either using close() method or using `with` clause

### The built-in `open(filename, mode)` function
- filename is the name of the file and may include path if the file is in a different location
- mode: specify mode in which you want to open a file. Example opening a file in read() cannot be used for writing, and vice versa.
- The `open()` function returns a file object.

### modes to open a file
1. r (read)
    - If the file doesn't exist it throws FileNotFoundError.
    - If the file exists we can read it
    - This mode is the default mode
2. w (write)
    - If the file doesn't exist it creates one.
    - If the file already exists then it erases all the existing data from it, to add new data to it.
3. a (append)
    - If the file doesn't exist it creates one.
    - If the file already exists then it appends the new data to the end of the file.
- **Note**: Both the write (w) and append (a) modes are used to write data to a file.
4. b (binary)
    - To read to a file file use 'rb' and to write to a binary use 'wb' modes

### The `r - read` mode
- This mode uses read(), readlines(), or readline() method

````python
# f_obj = open('practice-programs2.txt', 'r') # file doesn't exist so it throws an error
````

````python
f_obj = open("courses.txt", 'r')
print(f_obj.read())
f_obj.close()
````

### The `w - write` mode
- The write uses write() method to write the data to the file

````python
# opening a file that doesn't exist in write mode
f_obj = open("batch41-notes.txt", 'w')
f_obj.write("NOTES\n")
f_obj.write("========\n")
f_obj.write("Topics covered:\n")
f_obj.write("Introduction\nIf statements\nLoops\nLists\nTuples\nFiles\n")
f_obj.close()
````

````python
# opening an existing file in write erases the existing data and adds new data to the file
f = open("courses.txt", "w")
f.write("SKILLS TO MASTER\n")
f.write("-------------------\n")
f.write("Finance\nMarketing\nSales\nNegotation\nCommunication\nBusiness")
f.close()
````

### The `a - append` mode
- In the mode we use write() method

````python
# if the file doesn't exist then append mode creates one
f = open("power-bi-contents.txt", 'a')
f.write("Power-BI Core Concepts:\n")
f.write("Visualization\nDAX\nPower BI Service")
f.close()
````

````python
# if the file already exists then we can append new data to it in the append mode
f = open("power-bi-contents.txt", 'a')
f.write("\n\n")
f.write("The DAX is the hidden beast with layers in Power BI.\nWhich makes learning it difficult.")
f.close()
````

## 3 read methods

### `read()` method
- It reads entire file as one string

````python
# using read method
f = open("courses.txt", "r")
print(f.read())
f.close()
````

### `readlines()` method
- This method reads a file into a list, where each line from the file is an item in the list

````python
f = open("courses.txt", "r")
print(f.readlines())
f.close()
````

````python
f = open("courses.txt", "r")
print(f.readlines()[2])
f.close()
````

### `readline()` method
- Reads the next line starting from frist line

````python
f = open("courses.txt", "r")
line1 = f.readline()
line2 = f.readline()
line3 = f.readline()
print(line1, end='')
print(line2, end='')
print(line3, end='')
f.close()
````

### using `for` loop to read each line of a file

````python
f = open("courses.txt", "r")
for line in f:
    print(line, end='')
f.close()
````

### The `with` clause - contextual statment
- The `with` statement creates and context for opened file. Once the indented statements end the file closes automatically, without using close() method.
- Use `as` keyword to create an alias name for the file object

````python
with open("courses.txt", "r") as f:
    for line in f:
        print(line, end='')
````

## WORKING WITH CSV FILE

- The CSV stands for comma-separated values
- CSV files represent tabular data
- The csv file contains records, that are rows
- Each row values are separated by comma, and they are columns (fields)

##### Example
```
name,age,city,salary,dept
raju,35,hyd,28000.0,hr
vinod,40,blr,83000,sales
kumar,28,che,38000,admin
anand,29,amr,29000,marketing
```

- Each rows ends with a new line (\n) character
- Each row must have same number of columns

- To work with csv files, we need to import `csv` module and use function from it
- The `csv` module provides two function:
    1. `writer()` function to create a writer object so that we can write to the csv file. Additional we need to use `writerows(data)` method to write data to the writer object 
    2. `reader()` function to read from the csv file by creating reader object

**Note**: Use `newline=''` argument when opening a csv file using `open()` function

````python
emp_data = [
    ['name', 'city', 'dept', 'age', 'salary'],
    ['raju', 'hyd', 'hr', 35, 40000.0],
    ['kiran', 'amr', 'sales', 40, 80000.0],
    ['pradeep', 'blr', 'admin', 35, 63000.0]
]
````

````python
import csv

# writing data to the csv file
with open("emp_data.csv", "w", newline="") as f:
    w_obj = csv.writer(f)
    w_obj.writerows(emp_data)
````

````python
data = []
# reading data from the csv file
with open("emp_data.csv", "r", newline="") as f:
    r_obj = csv.reader(f)
    for line in r_obj:
        data.append(line)
print(data)
````

## WORKING WITH BINARY FILES

- A binary file is a non-text file
- It stores bytes

- Work with binary file, we need to use `pickle` module
- This module contains two functions:
    1. `dump()` to write to a file
    2. `load()` to read from a file

````python
emp_data = [
    ['name', 'city', 'dept', 'age', 'salary'],
    ['raju', 'hyd', 'hr', 35, 40000.0],
    ['kiran', 'amr', 'sales', 40, 80000.0],
    ['pradeep', 'blr', 'admin', 35, 63000.0]
]
````

````python
import pickle

# writing to a binary file
with open("employees.bin", "wb") as f:
    pickle.dump(emp_data, f)
````

````python
# reading from a binary file
with open("employees.bin", 'rb') as f:
    data = pickle.load(f)
    print(data)
````
