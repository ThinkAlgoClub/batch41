# WORKING WITH FILES

### Two types of files
1. Text files
2. Binary files

### Text files
- A text file contains a sequence of lines
- And each line contains characters and ends with a new line (\n) character
- Examples: txt, html, xml, json, csv, log, etc.

### Binary files
- A binary stores a sequence of bytes; may contain parts of text
- Not a text file
- Example: image, video, audio, compressed, database files, etc.

### Sequence of file operations
1. Open a file - open() function
2. Read from the file or write to the file - read() / write() methods
3. Close the file - close() method

**Note:** Always close the file; either using close() method or using `with` clause

### The built-in `open(filename, mode)` function
- filename is the name of the file and may include path if the file is in a different location
- mode: specify mode in which you want to open a file. Example opening a file in read() cannot be used for writing, and vice versa.
- The `open()` function returns a file object.

### modes to open a file
1. r (read)
    - If the file doesn't exist it throws FileNotFoundError.
    - If the file exists we can read it
    - This mode is the default mode
2. w (write)
    - If the file doesn't exist it creates one.
    - If the file already exists then it erases all the existing data from it, to add new data to it.
3. a (append)
    - If the file doesn't exist it creates one.
    - If the file already exists then it appends the new data to the end of the file.
- **Note**: Both the write (w) and append (a) modes are used to write data to a file.
4. b (binary)
    - To read to a file file use 'rb' and to write to a binary use 'wb' modes

### The `r - read` mode
- This mode uses read(), readlines(), or readline() method

```python
# f_obj = open('practice-programs2.txt', 'r') # file doesn't exist so it throws an error
```

```python
f_obj = open("courses.txt", 'r')
print(f_obj.read())
f_obj.close()
```

### The `w - write` mode
- The write uses write() method to write the data to the file

```python
# opening a file that doesn't exist in write mode
f_obj = open("batch41-notes.txt", 'w')
f_obj.write("NOTES\n")
f_obj.write("========\n")
f_obj.write("Topics covered:\n")
f_obj.write("Introduction\nIf statements\nLoops\nLists\nTuples\nFiles\n")
f_obj.close()
```

```python
# opening an existing file in write erases the existing data and adds new data to the file
f = open("courses.txt", "w")
f.write("SKILLS TO MASTER\n")
f.write("-------------------\n")
f.write("Finance\nMarketing\nSales\nNegotation\nCommunication\nBusiness")
f.close()
```

### The `a - append` mode
- In the mode we use write() method

```python
# if the file doesn't exist then append mode creates one
f = open("power-bi-contents.txt", 'a')
f.write("Power-BI Core Concepts:\n")
f.write("Visualization\nDAX\nPower BI Service")
f.close()
```

```python
# if the file already exists then we can append new data to it in the append mode
f = open("power-bi-contents.txt", 'a')
f.write("\n\n")
f.write("The DAX is the hidden beast with layers in Power BI.\nWhich makes learning it difficult.")
f.close()
```
