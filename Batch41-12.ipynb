# WORKING WITH DATES

##### Following examples are strings, not dates
```
'2026-06-17'
'06/17/2026'
'17-06-2026'
'06-17-2026'
```

- To create and work with date, time, and datetime values we need import `datetime` module
- The `datetime` module contains `date` constructor to construct date value, `time` constructor to construct time values, `datetime` constructor to construct datetime values

````python
# imports
from datetime import date
from datetime import time
from datetime import datetime
````

````python
# or import like the following
from datetime import date, time, datetime
````

### The `date(year, month, day)` constructor

- The date constructor accepts year, month, and day values to create a date object

````python
date(2026, 6, 17)
````

````python
type(date(2026, 6, 17))
````

````python
birth_day = date(2004, 7, 1)
````

````python
print(birth_day)
````

````python
# use today() method to get current day's date
e_roju = date.today()
print(e_roju)
````

### The `time(hours, minutes, seconds, microseconds)` construct

````python
login_time = time(9, 30)
print(login_time)
````

````python
logout_time = time(18, 42, 37, 1008)
print(logout_time)
````

````python
type(logout_time)
````

### `datetime(year, month, day [, hour][, minute][, second][, microsecond])` construct

````python
event = datetime(2026, 6, 30, 10, 30)
print(event)
````

````python
meeting_dt = datetime(2026, 8, 18, 17, 45, 30, 30000)
print(meeting_dt)
````

````python
# now() method of datetime construct returns current date and time
datetime.now()
````

````python
dt = datetime(2026, 12, 10)
````

````python
print(dt)
````
