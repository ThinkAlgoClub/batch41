### `datetime.strptime(datetime_str, format_str)` function
- The `strptime` method converts a string to datetime value
- The `p` in `strptime` stands for parsing

- `datetime_str` is the datetime/date/time value that is in string format
- `format_str` is the set of codes that helps us to format the `datetime_str`

#### datetime_str codes
- %y - two-digit year number
- %Y - four-digit year number
- %m - month number (1 - 12)
- %d - day number (1 - 31)
- %H - hour number (0 - 23)
- %I - hour number, 12 hours format (1 - 12)
- %M - minute number (0 - 59)
- %S - second number (0 - 59)
- %p - AM / PM specifier

````python
dt1 = "2026-12-31"
dt1 = datetime.strptime(dt1, "%Y-%m-%d")
print(dt1)
````

````python
dt2 = "28/5/2013"
dt2 = datetime.strptime(dt2, "%d/%m/%Y")
dt2
````

````python
dt3 = "On 24th of 7th month, in 2022, at 7:30 we had a meeting online"
dt3 = datetime.strptime(dt3, "On %dth of %mth month, in %Y, at %H:%M we had a meeting online")
dt3
````

````python
dt4 = "11:30 on 15/7/2023"
dt4 = datetime.strptime(dt4, "%H:%M on %d/%m/%Y")
dt4
````

### `datetime_odj.strftime(format_str)` method
- We can format a datetime object using `strftime` method
- The `f` in strftime stands for `format`

#### format_str codes
- %a - short weekday name (Sun, Mon, Tue, ...)
- %A - full weekday name (Sunday, Monday, Tuesday, ...)
- %b - short month name (Jan, Feb, Mar, ...)
- %B - full month name (Janurary, Februrary, March, ...)
- %Y - full year number
- %y - two digit year number
- %m - month number
- %d - day number
- %H - hour number (24-hour format)
- %I - hour number (12-hour format)
- %M - minute number
- %S - second number
- %p - AM / PM specifier

````python
meeting_dt = datetime(2026, 7, 15, 10, 30, 45, 1001)
````

````python
meeting_dt.strftime("(%A) %b %d, %Y")
````

````python
meeting_dt.strftime("%Y-%m-%d (%a)")
````

````python
meeting_dt.strftime("Date: %b %d, %Y - Time: %H:%M:%S")
````

````python
meeting_dt.strftime("Date: %b %d, %Y - Time: %H:%M:%S %p")
````

````python
meeting_dt.strftime("Date: %b %d, %Y - Time: %I:%M:%S %p")
````

````python
meeting_dt.strftime("%A %B %Y - %d ... %I:%M %p")
````

## WORKING WITH TIME DELTA (SPAN OF TIME)

- From `datetime` module import `timedelta` to create a timedelta object
- This object stores span of time or duration/period of time
- From `datetime` module import `timedelta` contructor to construct timedelta object

````python
from datetime import date, time, datetime, timedelta
````

### Syntax for `timedelta` constructor

```
timedelta([days, ][seconds, ][microseconds, ][weeks, ][hours, ][, minutes][, milliseconds])
```
- **Note:** Measuring units of timedelta are days, seconds, and microseconds

````python
duration = timedelta(weeks=6)
print(duration)
````

````python
project_duration = timedelta(weeks=14, days=2)
start_date = date(2026, 6, 22)
end_date = start_date + project_duration
print(end_date)
````

````python
task_time = timedelta(hours=3, minutes=30)
start_time = datetime.now()
end_time = task_time + start_time
print(end_time)
````

````python
print(type(end_date))
````

````python
print(type(task_time))
````

````python
print(type(project_duration))
````

### Finding difference between date values and datetime values also returns timedelta object

````python
today = date.today()
db_date = date(2026, 8, 17)
num_of_days = db_date - today
print(num_of_days)
````

````python
print(type(num_of_days))
````

````python
now = datetime.now()
event_date = datetime(2026, 9, 8, 11, 30)
waiting_time = event_date - now
print(waiting_time)
````
