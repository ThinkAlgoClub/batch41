# WORKING WITH DECIMALS

- Unlike floats, decimal number are exact
- To create a decimal number we need to use `Decimal` constructor from `decimal` module

```python
100.1 + 100.1 + 100.1
```

```python
from decimal import Decimal
```

```python
Decimal("100.1") + Decimal("100.1") + Decimal("100.1")
```

```python
print(Decimal("100.1") + Decimal("100.1") + Decimal("100.1"))
```

### **Rule:** Don't mix Decimals and Floats which is illegal. You can mix integers and Decimals

```python
# 100.1 + Decimal("100.1")   # type error
```

```python
100 + Decimal("100.1")
```

```python
print(Decimal("100.1") * 5)
```
