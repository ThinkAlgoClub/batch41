### 17. `format()` method

- It used like this: `str.format(dataitem(s))`
##### Syntax
```
"...{:format_specification}...".format(data_item(s))
```

#### `{:format_specification}` parts
- `[alignment][width][comma][.deci_places][type_code]`
- All the parts are in brackets in syntax, which means they are optional

#### Each part of the format_specification
- `alignment`: By default, text is alignment left, numeric values are aligned right.
    - '>' right-alignment
    - '<' left-alignment
    - '^' center-alignment
- `width`: space for the data_item
- `comma`: comma is for numbers
- `.deci_places`: .num_of_decimal_places can be specified
- `type_code`: `d` for integer; `f` for floats; `e` scientific notation

````python
"{:12}".format("hello")
````

````python
"{:10}".format(1234)
````

````python
"{:>12}".format("hello")
````

````python
"{:<10}".format(1234)
````

````python
"{:^10}".format(1234)
````

````python
x = 123456789765.8987654
"{:25,}".format(x)
````

````python
x = 123456789765.8987654
"{:25,.2f}".format(x)
````

````python
print("{:20} {:>10} {:>15}".format("PRODUCT NAME", "UNITS", "PRICE"))
print("{:20} {:>10} {:>15}".format("------------", "-----", "-----"))
print("{:20} {:10} {:15.3f}".format("HP Ryzen 7", 12, 78999.00))
print("{:20} {:10} {:15.3f}".format("Epson Projector", 3, 48000.00))
print("{:20} {:10} {:15.3f}".format("HP Inkjet Printer", 7, 18500.00))
print("{:20} {:10} {:15.3f}".format("Logitech Webcame", 2, 14599.00))
````

### 18. `f-strings`

- Allows to embeded variables and expressions directly into a string, by prefixing it with `f` and wrap the variables/expression in braces.

##### Syntax
```
f"...{expression}..."
```

````python
name = "Raju"
age = 35
print(f"My name is {name} and I am {age} years old.")
````

````python
pi = 3.14159
print(f"Pi rounded to two decimals {pi:.2f}")
````

````python
amount = 185000.0
print(f"The price of the product is: {amount:,.3f}")
````
