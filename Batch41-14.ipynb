### 8. `split([delimiter=' '])` method
- By default, splits a string at spaces, and returns a list with words
- However, we can explicitly specify the delimiter using the optional parameter

```python
quote = "How I wish I could calculate"
words = quote.split()
words
```

```python
dt = "2026/6/19"
date_v = dt.split()
date_v
```

```python
from datetime import date
dt = "2026/6/19"
date_values = dt.split("/")
year = int(date_values[0])
month = int(date_values[1])
day = int(date_values[2])
date(year, month, day)
```

### 11. `find(substring[, start_position])` method
- Finds the starting index of the substring in the larger string
- Optionally we can provide start_position using an index number to start the search
- If the substring is not found in the larger string then it returns -1

```python
email = "mahesh.kumar@gmail.com"
```

```python
at_index = email.find("@")
at_index
```

```python
first_dot_index = email.find(".")
first_dot_index
```

```python
dot_index = email.find(".", at_index + 1)
```

```python
if at_index > -1 and dot_index > at_index:
    print("Valid email address")
```

### 12. `rfind()` method
- Similar to find() method, but does search from right-side of the string

```python
email
```

```python
dot_index = email.rfind(".")
dot_index
```
