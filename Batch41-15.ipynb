# WORKING WITH DICTIONARIES

- A dictionary is a collection of unordered items
- In a dictionary the items are not indexed, so no index number, and it is efficient
- Every item in a dictionary has two components: key and value pairs
- We cannot have duplicate keys; however, two keys may be associated with the same value
- The keys are immutable types
- A dictionary is mutable

##### Syntax
```
{key1:value1, key2:value2, key3:value3, ...}
```

- Every item has a key and value pairs, each are separated by colon (:).

### Creating a dictionary

````python
numbers = {0:'zero', 1:'one', 2:'two', 3:'three', 4:'four'}
numbers
````

````python
# empty dictionary
catalogue = {}
````

### Accessing items from a dictionary
- It access an item from a dictionary use the `key` to get the value associated with that key
##### Syntax
```
dictionary_name[key]
```

````python
numbers
````

````python
numbers[0]
````

````python
numbers[3]
````

````python
courses = {1:"Computer Science", 2:"Information Technology", 3:"Electronics",
          4:"Civil", 5:"Mechanical"}
course_id = int(input("Enter course id: "))
print(courses[course_id])
````

##### Syntax


````python
courses
````

````python
print(courses)
````

### Adding new item to a dictionary
##### Syntax
```

```

````python
numbers
````

````python
numbers[5] = 'five'
numbers[6] = 'six'
print(numbers)
````

### Handling KeyError while accessing items from a dictionary

````python
countries = {"IN":"India", "JP":"Japan",
             "US":"United States", "UK":"United Kingdom"}
````

````python
countries["IN"]
````

````python
countries["JP"]
````

#### Using the `get()` we can avoid `KeyError`
##### Syntax
```
dict_name.get(key[, default_value])
```
- Provide the key to get associated value
- If the key isn't in the dictionary it gives None
- Additional, we can provide optional default value, instead of None

````python
countries
````

````python
countries["IN"]
````

````python
countries.get("US")
````

````python
countries.get("DE")   # returns None value
````

````python
countries.get("DE", "Unknown")
````

### Getting only keys from a dictionary

````python
# using for loop without any method
for key in countries:
    print(key, end=" ")
````

### Getting only keys from a dictionary using keys() method

````python
# Using keys() method to get keys from a dictionary
for key in countries.keys():
    print(key, end=" ")
````

### Getting only values from a dictionary using values() method

````python
# Using values() method to get values from a dictionary
for value in countries.values():
    print(value, end=", ")
````

### Getting key and value pairs as a tuple using items() method

````python
# Using items() method to get key and value pairs as a tuple from a dictionary
for item in countries.items():
    print(item)
````

````python
for key, value in countries.items():
    print("{:4} {:20}".format(key+":", value))
````

### Deleting items from a dictionary

````python
countries["AU"] = "Australia"
countries["CA"] = "Canada"
countries["DE"] = "Germany"
countries["RU"] = "Russia"
print(countries)
````

### Delete an item using `del` keyword

````python
print(countries)
````

````python
del countries["CA"]
print(countries)
````
