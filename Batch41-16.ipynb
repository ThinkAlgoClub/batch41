# WORKING WITH SETS

- A set is a collection of unordered items; items are not indexed (no index numbers)
- A set contains no duplicates
- A set is mutable

### Creating a set
##### Syntax
```
{item1, item2, item3, item4, ...}
```

````python
colors = {'red', 'green', 'blue', 'yellow', 'cyan', 'red', 'yellow'}
colors
````

### 1. `add(item)` method
- To add an item to a set

````python
colors
````

````python
colors.add('purple')
colors
````

````python
colors.add('black')
colors
````

### 2. ``update(list)` method
- To add a list of items to an existing set

````python
colors
````

````python
colors.update(['brown', 'white', 'orange', 'voilet', 'pink'])
print(colors)
````
