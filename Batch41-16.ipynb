# WORKING WITH SETS

- A set is a collection of unordered items; items are not indexed (no index numbers)
- A set contains no duplicates
- A set is mutable

### Creating a set
##### Syntax
```
{item1, item2, item3, item4, ...}
```

````python
colors = {'red', 'green', 'blue', 'yellow', 'cyan', 'red', 'yellow'}
colors
````

### 1. `add(item)` method
- To add an item to a set

````python
colors
````

````python
colors.add('purple')
colors
````

````python
colors.add('black')
colors
````

### 2. ``update(list)` method
- To add a list of items to an existing set

````python
colors
````

````python
colors.update(['brown', 'white', 'orange', 'voilet', 'pink'])
print(colors)
````

### 3. `remove(item)` method
- To remove the specified item from the set

````python
print(colors)
````

````python
colors.remove('red')
colors.remove('brown')
print(colors)
````

````python
# colors.remove('indigo') # KeyError
````

### 4. `discard(item)` item
- Removes the specified item from the set, if the is not in the set then it retunrs None

````python
print(colors)
````

````python
colors.remove('black')
````

````python
colors.discard('indigo')
````

````python
colors.discard('yellow')
````

````python
print(colors)
````

## SET OPERATIONS

### `union or |` - to combine two sets

````python
a = {1, 2, 3, 4, 5, 6}
b = {4, 5, 6, 7, 8, 9}
a.union(b)
a | b
````

### `intersection or &` - to get common elements from both the sets

````python
a = {1, 2, 3, 4, 5, 6}
b = {4, 5, 6, 7, 8, 9}
a.intersection(b)
a & b
````

### difference between two sets using `-` operator

````python
a = {1, 2, 3, 4, 5, 6}
b = {4, 5, 6, 7, 8, 9}
a - b
````

### `symmetric_difference or ^` for symmetric difference between sets

````python
a = {1, 2, 3, 4, 5, 6}
b = {4, 5, 6, 7, 8, 9}
a.symmetric_difference(b)
a ^ b
````

### `frozenset` - a set that cannot be changed

````python
a_set = frozenset([1, 2, 3, 4])
````

# PYTHON ARRAYS

- An array can contain only values of one type

- To create an array use array from the array module

````python
from array import array
````

````python
int_array = array('i', [10, 20, 30, 40, 50])
````

````python
float_array = array('f', [11.1, 12.2, 13.3, 14.4])
````
