# EXCEPTION HANDLING

## Error vs Exception

### Error
- It is syntactical (grammatical) mistake
- We need fix the error before running the code
- When there is an error we can't start executing the code

### Exception
- The code is valid (no errors)
- But something from outside can crash the program and terminates it
- We can't fix it, we need to handle the exceptions

````python
# can cause ValueError
value = int(input("Enter an integer :"))
print("Entered value is", value)
````

````python
# Can cause ZeroDivisionError
m = 10
n = 5
result = m / n
print(result)
````

### `try / except` construct to handle Exceptions

```python
# psuedo code
try:
    # code that might go wrong
    risky_operation()
except SomeException:
    # what if the specific exception occurs
    recovery_plan()
```

````python
# catching zero division error
try:
    numerator = 10
    denominator = 0
    result = numerator / denominator
    print("Result:", result)
except ZeroDivisionError:
    print("Oops! You cannot divide by zero.")
    
print("Program continues...")
````
