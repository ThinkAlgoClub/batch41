#### randint(min, max) function
- returns random a value between min and max values, both are inclusive

````python
random.randint(3, 5)
````

#### randrange(start, stop, step) function
- By default stop is 0, step is 1
- It returns one value between start and stop, excluding stop, values that are increment by step value

````python
random.randrange(3, 20, 7)
````

````python
random.randrange(3, 20, 7)
````

````python
random.randrange(3, 20, 7)
````
