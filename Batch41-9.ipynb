# WORKING WITH LISTS

- A list is a collection of ordered (sequential) items.
- A list contains items that are indexed. And the indexing starts from 0.
- Every has an index number starting from 0. First item's index is 0, second item's index is 1, third item's index is 2, so on and so forth.
- Similarly, we can refer items from the last by using negative indexing. Last item's index number is -1, second last item's index is -2, third last item's index number is -3, so on and so forth.
- List is mutable, it can change.

#### **Syntax**
```
[item1, item2, item3, item4, ...]
```

### 9. `sort([key=function, ][reverse=False])` method

- sorts a list in ascending or A-Z order by default
- The key parameter is used to specify a function before sorting a list
- The reverse parameter is used to specify the order, by default revers is False.

````python
cities = ['hyd', 'Del', 'blr', 'kol', 'che', 'ahm', 'bhu', 'png', 'mum', 'amr']
cities.sort()
print(cities)
````

````python
cities.sort(key=str.upper)
print(cities)
````

````python
cities.sort(key=str.lower, reverse=True)
print(cities)
````

### Shallow copy and Deep copy

##### shallow copy
- create a list by assigning it to a variable
- then assign that variable to another (new) variable
- now, both the variable point to the same list
- this is called shallow copying

````python
a = [1, 2, 3]
b = a
````

````python
print("List a:", a)
print("List b:", b)
````

````python
a.append(100)
print("List a:", a)
print("List b:", b)
````

````python
b[1] = 400
print("List b:", b)
print("List a:", a)
````

##### deep copy

- We can create a duplicate or another true copy of an existing list
- This is called deep copy
- To create a deep copy from `copy` module, use `deepcopy()` function

````python
import copy
l = [1, 2, 3]
m = copy.deepcopy(l)
print("List l:", l)
print("List m:", m)
````

````python
l.append(100)
print("List l:", l)
print("List m:", m)
````

````python
m[1] = 500
print("List m:", m)
print("List l:", l)
````

### `in` - a membership checker (keyword)
- We use `in` to check whether an item is in the list or not
- If the item is in the list it returns True; otherwise False
##### Syntax:
```
item in list
```

````python
print(cities)
````

````python
'mum' in cities
````

````python
'rch' in cities
````

````python
if 'jpr' in cities:
    pass # do nothing
else:
    cities.append('jpr')
````

````python
print(cities)
````

````python
city = 'Del'
if city not in cities:
    pass # do nothing
else:
    cities.remove(city)
````

````python
print(cities)
````

### List of lists
- A list containing items where each item is a list
- To access items we can use indexes, however, to access items from the inner lists use two indexes

````python
# matrix of 3 rows and 5 columns
student_marks = [
    ['rajesh', 100, 98, 77, 65],
    ['praveen', 70, 84, 92, 88],
    ['narsing', 99, 98, 97, 100]
]
````

````python
student_marks[0]
````

````python
student_marks[1]
````

````python
student_marks[1][0]  # use two indexes
````

````python
for student in student_marks:
    print(student)
````

````python
for student in student_marks:
    print(student[0])
````

````python
for student in student_marks:
    print(student[0], "-", student[4])
````
