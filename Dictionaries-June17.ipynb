# WORKING WITH DICTIONARIES

- A collection of items, unlike list items, they are unordered
- Each item in a dictionary contains `key` and `value` pairs
- In one sense, the keys word as indexes (in fact, no indexing here)
- The key values are unique in a dictionary, but values can repeat.
- The key values are of immutable type: int, float, str, tuple

##### **Syntax**
```
{key1:value1, key2:value2, key3:value3, key4:value4, ...}
```

#### Use `get()` to ignore KeyError
- `get(key[, DefaultValue])`
- If the key exists in the dictionary it returns the correspoding value
- If the key doesn't exists then it returns `None`
- We can also specify a default value if the key doesn't exist in the dictionary

````python
countries
````

````python
countries.get("IN")
````

````python
countries.get("DE")  # returns None
````

````python
countries.get("DE", "Unknown")
````

### Looping through keys

````python
countries
````

````python
for k in countries:
    print(k, end=' ')
````

````python
# Optionally we can also use - keys() method to get key values
for k in countries.keys():
    print(k, end=' ')
````

### Looping through values using `values()` method

````python
countries
````

````python
for v in countries.values():
    print(v, end=' ')
````

### Looping through items using `items()` method

- Returns a tuple of key and value pairs for each item

````python
for item in countries.items():
    print(item)
````

### Using `in` keyword to search for key in dictionary

````python
countries = {'IN': 'India',
 'US': 'United States',
 'RU': 'Russia',
 'JP': 'Japan',
 'UK': 'United Kingdom',
 'CA': 'Canada',
 'DE': 'Germany',
 'CH': 'China',
 'NP': 'Nepal',
 'AU': 'Australia'}
````

````python
'IN' in countries
````

````python
'IL' in countries
````

````python
key = 'MX'
if key in countries:
    pass   # do nothing
else:
    countries[key] = "Mexico"
````
