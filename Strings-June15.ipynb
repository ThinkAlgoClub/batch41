### 10. `find(substring [, start_position])` method
- Returns the index position of the first occurance of the specified substring

````python
at_index = email.find("@")
at_index
````

````python
dot_index_1st = email.find(".")
dot_index_1st
````

````python
dot_index = email.find(".", at_index + 1 )
dot_index
````

````python
dot_index = email.find(".", email.find("@") + 1)
dot_index
````

````python
if at_index > -1 and dot_index > at_index:
    print("This is an email address")
````

### 23. `format(dataitem(s))` method

##### **Syntax**
```
"...{:[width][comma][.deci_places][type]}...".format(dataitem)
```

#### `[width][comma][.deci_places][type]` - format specification
- `width` - optional; specifiy the width for datatime
- `comma` - optional; a number can contain comma (thousand separator)
- `.deci_places` - optional; how many decimal places are needed
- `type` - type of data you are formatting: d (for integers), f (for floats), e (for scientific notation)...
- `>` for right-alignment, `<` for left-alignment, `^` for center alignment

````python
"{:10}".format("hello")
````

````python
"{:15}".format(123)
````

````python
"{:20,}".format(12345645353634464545789)
````

````python
"{:20,.2f}".format(12345.67890987)
````

````python
"{:<20}".format(1000)
````

````python
"{:>20}".format("Hello")
````

````python
"{:^30}".format(12345)
````

````python
"{:20.4e}".format(123456789)
````

````python
print("{:25} {:>6} {:>20}".format("Product name", "Units", "Price"))
print("{:25} {:>6} {:>20}".format("------------", "-----", "-----"))
print("{:25} {:6} {:20,.2f}".format("Hp Ryzen 5", 10, 48000.0))
print("{:25} {:6} {:20,.2f}".format("Epson Projector", 2, 32500.0))
print("{:25} {:6} {:20,.2f}".format("Printer", 7, 18250.0))
print("{:25} {:6} {:20,.2f}".format("iPhone 17", 8, 180000.0))
````

````python
"My salary is {:,.3f} rupees.".format(100000.0)
````
