### Integer literals
- A number without a fractional part (decimal point)
- A complete number

- A number with a fractional part or decimal point
- An approximate number
- Examples:
    - 3.5, -8.001, 9.3, 3.14159, -9.8, 0.0, 7.0, 12.1, -1.3, -1.0

### Logical operators
- `not`, `and`, & `or` are logical operators
- Use these operators to combine Boolean expressions, and the result is a Boolean value
- Operator precedence: not, and, or

#### `not` operator
- The negation operator; it reverse the outcome of a Boolean expression
    - Negates True to False
    - Negates False to True
- *Syntax:*
```
    not bool_exp
```

````python
not True
````

````python
not False
````

````python
100 > 50
````

````python
not 100 > 50
````

````python
200 < 25
````

````python
not 200 < 25
````

#### `and` opertor
- Combines two Boolean expressions
- Returns True only when both the expressions are True; otherwise returns False
- *Syntax*
```
bool_exp1 and bool_exp2
```

````python
True and True
````

````python
True and False
````

````python
False and True
````

````python
False and False
````

````python
1 == 1 and 2 == 2
````

````python
1 == 1 and 2 != 2
````

````python
1 != 1 and 2 == 2
````

````python
1 != 1 and 2 != 2
````

#### `or` operator
- Combines two Boolean expressions
- Returns True when at least one Boolean expression is True; otherwise False
- *Syntax*
```
bool_exp1 or bool_exp2
```

````python
True or True
````

````python
True or False
````

````python
False or True
````

````python
False or False
````

````python
True and True and True and True and True
````

````python
True and True and False and True and True
````

````python
False or False or False or False or True or False or False
````
