# Comments
- Comments are written inside a code file, but they are not executed by the interpreter, instead comments are ignored by the interpreter
- In Python, we code comments starting with # hash tag, comments begin with hash tag (#)
- We can write comments on a separate line, or we can write those at the end of a statement

```python
# We are adding three values in this program

# values to add are stored in variables
x = 10    # value 1
y = 20    # value 2
z = 30    # value 3

# adding the three values
w = x + y + z   # addition

# pringing the result of the addition
print("Result:", w)   # print() statement
```

# Line Continuation character
- Sometimes we need to break a statement into multiple lines for the sake of readability
- In Python, `\' is a line-continuation character
- You type `\` and then `Return` key will continue the statement to next line

```python
1 + 10 + 2 + 20 + 3 + 30 + 4 + 40 + 5 + 50 + 6 + 60 + 7 + 70
```

```python
1 + 10 + \
2 + 20 + \
3 + 30 + \
4 + 40 + \
5 + 50 + \
6 + 60 + \
7 + 70
```
