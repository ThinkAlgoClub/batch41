## The `print([data][, sep=' '][, end="\n"])` function

- The function syntax has three arguments
- Each argument is in brackets - []
- Most of the time in the syntax, brackets [] imply optional part of the code (optional arguments)
- Since all the parameters are optional, if we don't use any parameter it prints empty line

````python
print()  # to print an empty line
````

- [data] -> optional, data can be one or more values separated by comma

````python
print(123)
````

````python
print(123,123)
````

````python
print(123,1.23,"Where there is a will there is a way!", 100 * 300, True)
````

````python
print(1,2,3,4,5)
````

````python
print(1, 2, 3, 4, 5)
````

````python
print(1,   2,   3,   4,        5)
````

- [sep=' '] - optional, but its default value is space (' ')
- It is used to separate arguments with text

````python
print(1, 2, 3, 4, 5, sep=' ')
````

````python
print(1, 2, 3, 4, 5, sep='-')
````

````python
print(1, 2, 3, 4, 5, sep='...')
````

````python
print(1, 2, 3, 4, 5, sep=' - ')
````

- [end="\n"] - optiona, but default value is \n
- \n is new line character; this takes the control to the next line
- The `end` argument prints its text after printing the last parameter

````python
print(1, 2, 3, 4, 5, end="...")
print(10, 20, 30, 40, 50, end="~~~")
print(100, 200, 300, 400, 500, end="...")
````

## The `input([prompt])` function

- Syntax:
```
var = input([prompt])
```
- The input() function waits for the user input
- Enter the input and hit Enter key
- It converts the input to a string
- Assigns the input value (which is converted to string) to the variable on the left-side

````python
x = 10
y = input()
print(type(x))
print(type(y))
````

````python
x = 10
y = input("Enter a numeric value: ")
print(type(x))
print(type(y))
````

````python
x = input("Enter an integer: ")
y = input("Enter another integer: ")
z = x + y
print("Result:", z)
````
