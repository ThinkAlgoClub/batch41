# The `if` statement / construct

- A conditional branching construct
- Each branch contains block (code)
- And block (code) is execute only from one branch or no-branch

- *Syntax:*
```
if bool_exp:
    block
[elif bool_exp:
    block]...
[else:
    block]
```

- The `if` statements has three clauses:
    - `if` clause: required; only one clause and first clause
    - `elif`clause: optional; 0 to n clauses
    - `else` clause: optional 0 to 1 clauses
- The `block` (code) is a part of clause
    - It is always indented
    - All the statements inside a block are equally indented
- The `bool_exp` returns True or False
    - Here when the clauses are evaluated (sequentially), the clause whose bool_exp is True first in the sequence the statements are executed from that clause (branch)
- The `else` statement execute only when all the above branches bool_exps are False.

````python
age = 17
if age >= 18:
   print("You can vote!")
   print("Voting is on June 30th, 2026")
   print("See you at the polling station.")
print("Goodbye")
````

````python
age = 3
if age >= 18:
   print("You can vote!")
   print("Voting is on June 30th, 2026")
   print("See you at the polling station.")
else:
    print("You can't vote now!")
    print("You can vote after", 18-age, "years.")
print("Goodbye")
````

````python
QUA = 'mba'
EXP = 3

input_qua = input("Enter your qualification: ")
input_exp = int(input("Enter your experience in years: "))

if input_exp >= EXP:
    if input_qua == QUA:
        print("You are eligible to attend the interview.")
    else:
        print("You have experience but no qualification.")
else:
    if input_qua == QUA:
        print("You are qualified but not experienced.")
    else:
        print("You are neither qualified nor experienced.")
````

````python
QUA = 'mba'
EXP = 3

input_qua = input("Enter your qualification: ")
input_exp = int(input("Enter your experience in years: "))

if input_exp >= EXP and input_qua == QUA:
        print("You are eligible to attend the interview.")
elif input_exp >= EXP and input_qua != QUA:
        print("You have experience but no qualification.")
elif input_exp < EXP and input_qua == QUA:
        print("You are qualified but not experienced.")
else:
        print("You are neither qualified nor experienced.")
````
