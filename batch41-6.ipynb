# Loop Constructs

- A loop structure / construct is intended to repeatedly execute a block
- There are two types of loops:
    1. Collection / countable loops - for loop
    2. Conditional loops - while loop

### Collection / countable loop - for

- The `for` loop is collection loop
- If a collection has n items, then `for` loop executes its block for n-times
- This is a finite loop

- *Syntax:*
```
for <var> in <collection>:
    block
```

- collection: the collection of items/objects
- For each collection the `block` executes once
- Each execution is called an `iteration`
- In each iteration, an item from the collection is assigned to the `<var>`

````python
for i in [85, -10, 73]:
    print("Hello, World!")
    print("~" * 15)
````

````python
for name in ["Vineeth", "Srinivas", "Aarthi", "Joyce", "Supriya"]:
    print("Employee:", name)
    print("Welcome to the DataTeach.ai")
    print()
````

````python
for emp in ("1001", "Rajesh", 35, 80000.0, True, 'Hyd', 'HR', 8.3):
    print("->", end=" ")
    print(emp)
````

````python
for l in "INDIA":
    print(l)
````

````python
for l in "INDIA":
    print(l, end=" ")
````

````python
total = 0
for i in [1, 2, 3, 4, 5]:
    total = total + i   # total += i
print(total)
````

## The `range([start,] stop[, step])` iterator function
- The range() returns next value from the series that it generates for each iteration of the loop.
- *Syntax*
```
range([start,] stop[, step])
```

- `[start]` - optional; default is 0; its inclusive
- `stop` - required; its exclusive
- `[step]` - optional; default is 1
- The `range()` function returns series values for each iteration starting from `start` value till `stop` value, which is exclusive, by adding `step` till it reaches `stop` value.

#### range(stop)
- start is 0; step is 1; stop is +ve

- range(5)   # 0, 1, 2, 3, 4
- range(3)   # 0, 1, 2
- range(8)   # 0, 1, 2, 3, 4, 5, 6, 7

````python
for i in range(5):
    print(i, end=" ")
````

````python
for i in range(8):
    print(i, end=" ")
````

````python
for i in range(3):
    print(i, end=" ")
````

### range(start, stop)
- step is 1; start < stop

- range(1, 5)    # 1, 2, 3, 4
- range(10, 17)  # 10, 11, 12, 13, 14, 15, 16
- range(-3, 3)   # -3, -2, -1, 0, 1, 2
- range(-7, -1)  # -7, -6, -5, -4, -3, -2
- range(3, -3)   # WRONG

````python
for i in range(1, 5):
    print(i, end=" ")
````

````python
for i in range(10, 17):
    print(i, end=" ")
````

````python
for i in range(-3, 3):
    print(i, end=" ")
````

````python
for i in range(-7, -1):
    print(i, end=" ")
````

````python
for i in range(3, -3):
    print(i, end=" ")
````

### range(start, stop, step)
- start < stop; step is +ve
- start > stop: step is -ve

- range(10, 100, 20)   # 10, 30, 50, 70, 90
- range(-100, 100, 40) # -100, -60, -20, 20, 60
- range(100, -100, 30) # WRONG
- range(100, -100, -30) # 100, 70, 40, 10, -20, -50, -80
- range(10, 30, 4)  # 10, 14, 18, 22, 26
- range(-10, -20, -3)  # -10, -13, 16, -19

````python
for i in range(-10, -20, -3):
    print(i, end=" ")
````

## Conditional loop - `while`
- This loop iterates as long as the bool_exp is True; once the bool_exp is False it stops executing (iterating)
- *Syntax*
```
while bool_exp:
    block
```


````python
# infinite loops
while 1 < 5:
    print("Hello, World!")

while True:
    print("Goodbye")
````

````python
# loops that never run
while False:
    print("Good morning!")

while 1 > 5:
    print("Hello, Pythonista!!")
````

````python
# printing 0 to 4 using `for` loop
# start is 0, stop is 5, step is 1
for i in range(5):
    print(i, end=' ')
````

````python
# priing 0 to 4 using `while` loop
i = 0   # start is 0
while i < 5:   # stop is 5
    print(i, end=' ')
    i += 1     # step is 1
````

````python
# while to create an interactive program
print("QUOTE TO TITLE TEXT")
run = 'y'

while run.lower() == 'y':
    quote = input('Enter a quote: ')
    print(quote.title())
    run = input("Would like to convert another quote to title case (y/n): ")

print()
print("Goodbye!")
````

### The `break` statement
- Using `break` we can end (terminate) a loop early, without going through all the iterations
- We need to break a loop for a reason, and the reason is specified using `if` statement
- Always use the `break` statement inside a `if` statement

````python
score = [99, 100, 78, 85, 108, 74, 99, 45, 67, 70, 77, 84, 61]
for i in score:
    if i > 100:
        print("\nUnusual value:", i)
        break
    print(i, end=' ')
````

````python
score = [99, 100, 78, 85, 108, 74, 99, 45, 67, 70, 77, 84, 61] # value divisible by 15 then break
for i in score:
    if i % 15 == 0:
        print(i, end=' ')
        break
    print("-", end=" ")
````

### The `continue` statement
- To skip an iteration we use the `continue` statement
- The continue statement takes the control to the beginning of the loop immediately, without executing other following statements within the loop

````python
score = [99, 100, 78, 85, 108, 74, 99, 45, 67, 70, 77, 84, 61]
for i in score:
    if i % 3 != 0:
        continue
    print(i, end=' ')
````

````python
score = [99, 100, 78, 85, 108, 74, 99, 45, 67, 70, 77, 84, 61]
for i in score:
    if i % 3 == 0:
        continue
    print(i, end=' ')
````
