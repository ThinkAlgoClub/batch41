# User-Defined Functions (UDFs)

- To write a reusable code
- To make your code manageable
- To make your code easy to debug and update

#### **Syntax:**
```
def func_name([args]):
    block
```

- Use `def` keyword to define a function
- `func_name` is the name that we give to function, follows the same naming convention that applies to variables
- `[args]` - arguments; these are inputs a function expects; optional
- `block` - indented statements under the `def` clause
- **Calling a function**: Function definition doesn't execute a function, we need to call functions explicitly; type function, and inside parentheses code the required arguments

````python
# defining a function; without arguments (parameters)
def welcome():
    print("Welcome to world of Python!")
    print("Today we are learning UDFs!!")
    print("Give me a week's time to master functions!!!")

# function call
welcome()
````

````python
# defined a function
def welcome(name):
    print("Hello,", name)
    print("Welcome to the world of Python coding!")

# calling the function
# welcome()
````

````python
def welcome(name):
    print("Hello,", name)
    print("Welcome to the world of Python coding!")

welcome("Mani")
````

````python
def welcome(name, age):
    print("Hello,", name)
    print("Your age:", age)

# welcome("Anji Reddy")
````

````python
def welcome(name, age):
    print("Hello,", name)
    print("Your age:", age)

welcome(35, "Anji Reddy")
````

### Requried to optional arguments

- To make an argument optional, we need set a value to that argument in the first line of the function defintion (within parentheses); and the argument becomes optional with a default value.
- When we provide a argument it overwrites the default-value; otherwise it takes default value.
- Code required arguments first, then code optional arguments with default value.

````python
# function with one parameter which is optional with default-value
def welcome(name='patron'):
    print("Hello,", name)

welcome()
````

````python
def product(name, price, discountPercent=2):
    print("Name:", name)
    print("Price:", price)
    print("Discount percent:", discountPercent)
    print("Discount amount:", price * discountPercent / 100)
    print("Discount price:", price * (1 - discountPercent/100) )

product("Epson Projector", 40000.0, 5)
````

### Name paramenters to overwrite positional arguments

- In the function we can use the names of the parameters to pass the values to the function
- That makes sense when we don't remember the order of the parameters

````python
def product(name, price, discountPercent=2):
    print("Name:", name)
    print("Price:", price)
    print("Discount percent:", discountPercent)
    print("Discount amount:", price * discountPercent / 100)
    print("Discount price:", price * (1 - discountPercent/100) )

product(price=40000.0, name="Epson Projector")
````

### The `return` statement

- The `return` statement returns a value(s)/object(s) from the function to the calling statement
- The `return` statement ends the function execution

````python
# function without 'return' statement
def adding(a, b):
    c = a + b
    print("Result:", c)

adding(5, 10)
````

````python
# using return statement
def adding(a, b):
    c = a + b
    return c

adding(10, 20)
````

````python
def adding(a, b):
    c = a + b
    return c
    print("Hello, Dilip Kumar!!")
    print("Good morning!!")

result = adding(10, 20)
print("Result:", result)
````

````python
def perfect_number(n):
    total = 0
    for i in range(1, n):
        if n % i == 0:
            total += i
    if total == n:
        return "Perfect number"
    else:
        return "Not a perfect number"

perfect_number(500)
````

### Global and Local variables

#### **Global**
- These are declared or created outside of any function
- They are accessible everywhere, outside functions and inside any function
- By default, global variables are read-only inside a function
- We can modify the value of a global variable inside a function using the `global` keyword, which is not recommended
- If a local and global variable has same name, then within a function local variable gets precedence
#### **Local**
- These variable are declared or created inside a function
- And their scope is limited to those function, they are not avaiable outside of that specific function

````python
def sample():
    print(x)   # global variable

x = 25        # global variable
sample()
````

````python
def sample():
    x = 30     # local variable
    print(x)   # local variable

x = 10        # global variable
sample()
print(x)      # global variable
````

````python
def sample(x):  # local here
    x = 2 * x
    print(x)

x = 30      # global variable
sample(x)   # global variable
print(x)    # global variable
````

````python
def sample():  # local here
    global x   # not recommeneded to make a global variable writable
    x = 2 * x
    print(x)

x = 30      # global variable
sample()   # global variable
print(x)    # global variable
````

````python
TAX = 0.2

def tax_amount(amount):
    return amount * TAX

def paid_amount(amount):
    return amount * ( 1 - TAX)

salary = 60000
print("Tax:", tax_amount(salary))
print("Salary:", paid_amount(salary))
````

## Modules and Functions

- A python file is also called a python module
- A python file name `temperature.py` is called module name `temperature`
- Generally, a python module contains functions related to a specific category

#### Using a module and its function
- Import a module to use it:
```
import module_name

import temperature
```

- Then to use a function from the module, use the module_name as a prefix to the function_name:
```
module_name.function_name([args])
temperature.to_celsius(100)
```

````python
import temperature
temperature.to_celsius(100)
````

````python
temperature.to_fahrenheit(30)
````

#### Use alias name to load a module

````python
import temperature as t   # as keyword is used for aliasing
t.to_celsius(120)
````
